## Fill NA values with mean and mode values

```python
df1=df.copy()
for col in df.columns:
    if df1[col].isnull().any():
        if df1[col].dtype == 'object':  # For categorical (non-numeric) columns
            df1[col] = df1[col].fillna(value=df1[col].mode()[0])
        else:  # For numeric columns
            df1[col] = df1[col].fillna(value=df1[col].mean())

# Check for remaining missing values
df1.isnull().sum()
df1
```

## Fill NA values with forward fill values

```python
df2=df.copy()
for col in df2.columns:
    if df2[col].isnull().any():
        df2[col] = df2[col].ffill()
df2.isnull().sum()
df2
```

## Fill NA values with backward fill values

```python
df3=df.copy()
for col in df3.columns:
    if df3[col].isnull().any():
        df3[col] = df3[col].bfill()
df3.isnull().sum()
df3
```
