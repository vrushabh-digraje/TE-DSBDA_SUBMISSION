```python
df.isnull().sum()
```

## Fill NA values with mean and mode values

```python
df1=df.copy()
for col in df.columns:
    if df1[col].isnull().any():
        if df1[col].dtype == 'object':  # For categorical (non-numeric) columns
            df1[col] = df1[col].fillna(value=df1[col].mode()[0])
        else:  # For numeric columns
            df1[col] = df1[col].fillna(value=df1[col].mean())

# Check for remaining missing values
df1.isnull().sum()
df1
```
